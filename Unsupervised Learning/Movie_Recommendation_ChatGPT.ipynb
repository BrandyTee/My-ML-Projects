```python
#Import Dependencies
import numpy as np
import pandas as pd
import difflib
from google.colab import drive
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
drive.mount("/content/drive")
```

```python
path = "/content/drive/MyDrive/datasets/movies recommendation.csv"
movies_data = pd.read_csv(path)
movies_data.head()
```

```python
movies_data.info()
```

```python
movies_data.shape
```

```python
X = movies_data.iloc[:15000, :]
X.tail()
```

```python
#Select Relevant Features
relevant_features = ["Mystery", "Sci-Fi", "Thriller", "Action", "Adventure", "Animation", "Children", "Comedy", "Fantasy", "Romance"]
print(relevant_features)
```

```python
#Combining the relevant_features
combined_features = X["genres"] + " " + X["title"]
```

```python
print(combined_features)
```

```python
#Convert Textual Data To Features
vectorizer = TfidfVectorizer()

feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
#Use cosine similarity to get similarity scores
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
#Getting the movie name
movie_name = input("Enter Movie: ")
```

```python
#List of all movies
all_movies = X["title"].tolist()
print(all_movies)
```

```python
#Finding the close match for movie name
find_close_match = difflib.get_close_matches(
movie_name, all_movies
)
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
#Find the index of the movie
index_movie = X[X.title == close_match].index[0]
print(index_movie)
```

```python
similarity_score = list(enumerate(similarity[index_movie]))
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
#Print the names of similar movies based on the index

print('Movies Suggested for you: \n')

i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if  ( i < 20):
        print(i, '.', title_from_index)
        i += 1
```
